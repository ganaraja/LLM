# BM25 Keyword Search

```python
import bm25s
from sklearn.feature_extraction import _stop_words
import string

corpus = [

"The solar system consists of the Sun and the celestial objects that orbit it. These include planets, moons, \
asteroids, comets, and meteoroids. The Sun is the central star, providing the gravitational force that keeps \
the system together. The eight planets include Mercury, Venus, Earth, Mars, Jupiter, Saturn, Uranus, and Neptune.",

"Machine learning is a branch of artificial intelligence that enables systems to learn and make decisions without \
being explicitly programmed. It involves the use of algorithms and statistical models to identify patterns in data.\
Common applications include image recognition, natural language processing, and predictive analytics.",

"World War II was a global conflict that lasted from 1939 to 1945. It involved most of the world's nations and \
was marked by significant events such as the Holocaust, the atomic bombings of Hiroshima and Nagasaki, and the \
D-Day invasion. The war ended with the Allied victory and the establishment of the United Nations.",

"Python is a high-level, interpreted programming language known for its readability and simplicity. It supports \
multiple programming paradigms, including procedural, object-oriented, and functional programming. Popular libraries \
include NumPy, pandas, and TensorFlow, making Python widely used in data science and machine learning.",

"The Eiffel Tower is a wrought-iron lattice tower located in Paris, France. It was designed by Gustave Eiffel and \
completed in 1889 as the entrance arch to the 1889 World's Fair. Standing 330 meters tall, it is an iconic symbol \
of France and one of the most visited landmarks in the world.",

"Climate change refers to long-term changes in temperature, precipitation, and other atmospheric conditions on Earth.\
It is primarily driven by human activities, such as burning fossil fuels and deforestation. The effects of climate \
change include rising sea levels, extreme weather events, and loss of biodiversity.",

"Mathematics has evolved over thousands of years, starting with ancient civilizations like the Babylonians and \
Egyptians. Key contributions include the development of geometry by the Greeks, the introduction of zero by Indian \
mathematicians, and the formulation of calculus by Newton and Leibniz.",

"The Internet of Things refers to a network of physical devices connected to the internet, enabling them to collect \
and exchange data. IoT applications include smart homes, industrial automation, and healthcare monitoring. Challenges \
include security, privacy, and interoperability.",

"William Shakespeare was an English playwright, poet, and actor. His famous works include tragedies like Hamlet, Macbeth, \
and Othello, as well as comedies like 'A Midsummer Night's Dream' and 'Twelfth Night'. His writing has had a profound \
influence on the English language and literature.",

"The human brain is a complex organ responsible for processing sensory information, controlling motor functions, and \
enabling cognition. It consists of the cerebrum, cerebellum, and brainstem. The brain's neurons communicate through \
electrical and chemical signals, forming intricate neural networks.",

]
```

```python
# We lower case our text and remove stop-words from indexing
def bm25_tokenizer(text):
    tokenized_doc = []
    for token in text.lower().split():
        token = token.strip(string.punctuation)

        if len(token) > 0 and token not in _stop_words.ENGLISH_STOP_WORDS:
            tokenized_doc.append(token)
    return tokenized_doc

# Tokenize the corpus and index it
corpus_tokens = bm25s.tokenize(corpus)
retriever = bm25s.BM25(corpus=corpus)
retriever.index(corpus_tokens)
```

```python
# This function will search all docs to answer the query
def search(query):
    print("\nInput question:", query)

    tokenized_query = query.split()

    docs, _ = retriever.retrieve([tokenized_query], k=3)
    scores = retriever.get_scores(tokenized_query)
    print(f"\nScores by documents: {scores}")
    print("\nTop n=3 Documents:\n")
    print(docs)
```

```python
search(query = "machine learning applications")
```

```python
search(query = "famous landmarks in France")
```

### ***Just a contrast with semantic search***

```python
search(query = "tell me something about galaxy or astronomy")
```

```python
# For Semantic Search
from sentence_transformers import SentenceTransformer
from sentence_transformers import util

MODEL = 'msmarco-distilbert-base-v4'
embedder = SentenceTransformer(MODEL)

embeddings = embedder.encode(corpus, convert_to_tensor=True)
```

```python
query_text = "tell me something about galaxy or astronomy"
query = embedder.encode(query_text, convert_to_tensor=True)

search_results = util.semantic_search(query, embeddings, top_k = 3)
search_results
```

```python
for index, result in enumerate(search_results[0]):
    print('-'*80)
    print(f'Search Rank: {index}, Relevance score: {result["score"]} ')
    print(corpus[result['corpus_id']])
```

### ***Example where BM25's keyword search trumps a bit***

```python
search(query = "english")
```

```python
query_text = "english"
query = embedder.encode(query_text, convert_to_tensor=True)

search_results = util.semantic_search(query, embeddings, top_k = 3)
search_results
```
